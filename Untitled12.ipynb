```python
import os
from typing import Literal
from typing_extensions import TypedDict
from dotenv import load_dotenv
from google.colab import userdata

# Install required packages if not already installed
!pip install -U langchain-google-genai langchain-community python-dotenv langgraph fpdf2 pypdf

# LangChain & LangGraph Imports
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.document_loaders import PyPDFLoader
from langgraph.graph import StateGraph, START, END

# Load environment variables (for local development) - Note: Colab userdata is preferred
load_dotenv()

# IMPORTANT: Retrieve your API key securely from Colab's user data.
# Go to the left panel (key icon) -> Secrets -> Add a new secret named 'GEMINI_API_KEY'.
GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')

# ==========================================
# 1. Define the Shared State
# ==========================================
class AgentState(TypedDict):
    pdf_path: str   # Path to the input PDF file
    question: str   # User query
    category: str   # 'on_topic' or 'off_topic'
    context: str    # Will hold the extracted text from the PDF
    answer: str     # Final generated response

# Initialize Gemini with the API key
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0, google_api_key=GEMINI_API_KEY)

# ==========================================
# 2. Define the Graph Nodes (The Actions)
# ==========================================

def load_pdf_context(state: AgentState) -> dict:
    """NODE 0: Reads the PDF file path from the state and extracts all text content."""
    print(f"--- NODE 0: Extracting text from PDF: {state['pdf_path']} ---")

    if not os.path.exists(state["pdf_path"]):
        raise FileNotFoundError(f"The PDF file at {state['pdf_path']} was not found. Please create it first.")

    loader = PyPDFLoader(state["pdf_path"])
    docs = loader.load()

    # Combine the content of all pages into one comprehensive context string
    full_pdf_text = "\n".join([doc.page_content for doc in docs])

    # Update the graph state with the dynamic context
    return {"context": full_pdf_text}


def classify_input(state: AgentState) -> dict:
    """NODE 1: Evaluates if the question is relevant to the extracted PDF context."""
    print("--- NODE 1: Classifying Input against PDF Context ---")

    prompt = ChatPromptTemplate.from_messages([
        ("system", (
            "You are a routing assistant. Analyze the provided PDF Context and the User Question.\n"
            "Classify the query into exactly one of these categories:\n"
            "- 'on_topic': If the user's question asks about information, data, or concepts explicitly discussed in the PDF context.\n"
            "- 'off_topic': If the question is completely unrelated to the PDF content.\n\n"
            "Reply with ONLY the category name ('on_topic' or 'off_topic').\n\n"
            "PDF Context:\n{context}"
        )),
        ("user", "{question}")
    ])

    chain = prompt | llm | StrOutputParser()
    category = chain.invoke({"question": state["question"], "context": state["context"]}).strip().lower()

    return {"category": category}


def retrieve_and_answer(state: AgentState) -> dict:
    """NODE 2a: Executes RAG by answering the question strictly using the PDF context."""
    print("--- NODE 2a: Executing RAG (Answering from PDF) ---")

    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are an expert document assistant. Answer the question using ONLY the provided PDF context. If the answer cannot be found in the context, say 'I cannot find that information in the document.'\n\nPDF Context:\n{context}"),
        ("user", "{question}")
    ])

    chain = prompt | llm | StrOutputParser()
    answer = chain.invoke({"question": state["question"], "context": state["context"]})

    return {"answer": answer}


def handle_off_topic(state: AgentState) -> dict:
    """NODE 2b: Handles off-topic queries gracefully."""
    print("--- NODE 2b: Handling Off-Topic Query ---")

    answer = "I'm sorry, I am guardrailed to only answer questions directly pertaining to the uploaded PDF document."
    return {"answer": answer}

# ==========================================
# 3. Define the Router (The Conditional Logic)
# ==========================================

def route_question(state: AgentState) -> Literal["retrieve_and_answer", "handle_off_topic"]:
    """Determines the next path depending on whether the query fits the PDF content."""
    if "on_topic" in state["category"]:
        return "retrieve_and_answer"
    else:
        return "handle_off_topic"

# ==========================================
# 4. Build the Graph Workflow
# ==========================================

workflow = StateGraph(AgentState)

# Register all 4 nodes
workflow.add_node("load_pdf_context", load_pdf_context)
workflow.add_node("classify_input", classify_input)
workflow.add_node("retrieve_and_answer", retrieve_and_answer)
workflow.add_node("handle_off_topic", handle_off_topic)

# Map execution sequences
workflow.add_edge(START, "load_pdf_context")          # Start by parsing the PDF
workflow.add_edge("load_pdf_context", "classify_input") # Move to classification

# Router step after classification
workflow.add_conditional_edges(
    "classify_input",
    route_question,
    {
        "retrieve_and_answer": "retrieve_and_answer",
        "handle_off_topic": "handle_off_topic"
    }
)

# Connect endpoints to final termination marker
workflow.add_edge("retrieve_and_answer", END)
workflow.add_edge("handle_off_topic", END)

app = workflow.compile()

# ==========================================
# 5. Run Execution Tests
# ==========================================
if __name__ == "__main__":
    sample_pdf_path = "/content/S Santhosh Kumar Resume.pdf"

    # Quick guard check to make sure the PDF exists before running the loop
    if not os.path.exists(sample_pdf_path):
        print(f"Creating a sample PDF '{sample_pdf_path}' for testing...")
        from fpdf import FPDF
        try:
            pdf = FPDF()
            pdf.add_page()
            pdf.set_font("Arial", size=12)
            pdf.cell(200, 10, txt="ACME Corp Policy Document", ln=1, align="C")
            pdf.cell(200, 10, txt="1. Remote Work: Employees can work remotely up to 3 days a week.", ln=2)
            pdf.cell(200, 10, txt="2. Dress Code: Smart casual is required from Monday to Thursday. Fridays are casual.", ln=3)
            pdf.output(sample_pdf_path)
        except ImportError:
            print("Note: Please ensure 'sample_company_policy.pdf' exists in your directory.")

    print("\n==================================================")
    print("🤖 Gemini RAG Agent Initialized!")
    print("Type your questions below. Type 'stop' to exit.")
    print("==================================================\n")

    # Start the continuous loop
    while True:
        inpt = input("Enter your Query: ")

        # 1. Check for the stop condition immediately (case-insensitive)
        if inpt.strip().lower() == 'stop':
            print("\nExiting chat. Goodbye!")
            break

        # 2. Handle accidental empty presses gracefully
        if not inpt.strip():
            print("Please enter a valid question or type 'stop'.\n")
            continue

        # 3. Construct the fresh payload for this specific turn
        inputs = {
            "pdf_path": sample_pdf_path,
            "question": inpt
        }

        # 4. Invoke the LangGraph application and display the output
        result = app.invoke(inputs)
        print(f"\nFinal Answer:\n{result['answer']}")
        print("\n" + "-"*40 + "\n")
```
